```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ms")

text = "You are amazing!"
tokenized_input = tokenizer.tokenize(text)
tokenized_input
```

```python
model_name ='Helsinki-NLP/opus-mt-en-ms'

model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

def translate_text(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors = 'pt', padding = True, truncation = True)

    translation_ids = model.generate(
        input_ids,
        max_length = 200,
        num_beams = 5,
        length_penalty = 1.0,
        no_repeat_ngram_size = 2,
        top_k = 50,
        top_p = 0.95,
        early_stopping = False,
        do_sample = True
    )

    translated_text = tokenizer.decode(translation_ids[0], skip_special_tokens = True)

    return translated_text

input_text = "私はこの翻訳アプリをテストするためにこの文を書いています、テストテスト、こんにちは、何をしていますか?、このゲームは楽しいですか?"
translated_text = translate_text(input_text)
translated_text
```
